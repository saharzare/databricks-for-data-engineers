```python
%sql
SELECT
    s.product_number   AS sales_pn,
    p.product_number   AS gold_pn
FROM workspace.silver.crm_sales s
LEFT JOIN workspace.gold.dim_production p
  ON s.product_number = p.product_number
WHERE p.product_number IS NOT NULL
LIMIT 10;
```

```python
%sql
-- محصولاتی که در sales هستن ولی در dim_production نیستن
SELECT COUNT(*) AS not_matched_products
FROM workspace.silver.crm_sales s
LEFT ANTI JOIN workspace.gold.dim_production p
  ON s.product_number = p.product_number;
```

```python
%sql
SELECT
  COUNT(*) AS total_rows,
  SUM(CASE WHEN p.product_key IS NOT NULL THEN 1 ELSE 0 END) AS matched_rows,
  ROUND(
    100.0 * SUM(CASE WHEN p.product_key IS NOT NULL THEN 1 ELSE 0 END) / COUNT(*),
    2
  ) AS match_rate_percent
FROM workspace.silver.crm_sales s
LEFT JOIN workspace.gold.dim_production p
  ON s.product_number = p.product_number;
```

##Reading silver

```python
##reading 
query = """ SELECT s.*, p.product_key,c.customer_key FROM workspace.silver.crm_sales as s left join workspace.gold.dim_production as p on s.product_number = p.product_number left join workspace.gold.dim_customers as c on s.customer_id = c.customer_id """ 
df = spark.sql(query) 
display(df )
```
